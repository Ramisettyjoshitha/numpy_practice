```python
import numpy as np
```

Create an array containing numbers 1 to 12. Reshape it into a 3 × 4 array. Find its shape, ndim, and size.

```python
a=np.arange(1,13)
b=a.reshape(3,4)
print(b)
print(b.shape)
print(b.size)
print(b.ndim)
```

Create an array containing numbers 1 to 24. Reshape it using (4, -1). Predict the final shape before running it.

```python
w=np.arange(1,25)
v=w.reshape(4,-1)
print(v.shape)
print(v)
```

Create this array:
[[1, 2, 3],
 [4, 5, 6]]

Predict the output of reshaping it to (3, 2), then predict the output of transposing it. Explain why the results differ.

```python
i=np.array([[1, 2, 3],
 [4, 5, 6]])
n=i.reshape(3,2)
print(n)

print(n.T)
```

Convert a 3 × 4 array into a 1D array using both flatten() and ravel(). State the main difference between them.

```python
l=np.arange(1,13).reshape(3,4)
o=l.flatten()
k=o.ravel()
print(o)
print(k)
```

A 1D array has shape (5,). Convert it first to shape (1, 5) and then separately to shape (5, 1).

```python
a = np.array([1, 2, 3, 4, 5])
row=a[np.newaxis,:]
col=a[:,np.newaxis]
print(row.shape)
print(col.shape)
```

n array has shape (1, 3, 1). What will its shape become after using squeeze()

```python
a = np.array([[[1], [2], [3]]])
b=a.squeeze()
print(b.shape)
```

For this array:
[[1, 2, 3],
 [4, 5, 6]]

Find the sum using axis=0 and axis=1. Predict both outputs before running the code.

```python
a=np.array([[1, 2, 3],
 [4, 5, 6]])
b=a.sum(axis=0)
c=a.sum(axis=1)
print(b)
print(c)
```

For this array:
[[10, 20],
 [30, 40],
 [50, 60]]

Find the mean using axis=0 and axis=1

```python
a=np.array([[10, 20],
 [30, 40],
 [50, 60]])
b=a.mean(axis=0)
c=a.mean(axis=1)
print(b)
print(c)
```
