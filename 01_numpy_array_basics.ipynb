Question 1: Create a 1D array containing 10, 20, 30, 40, 50. Find its ndim, shape, size, and dtype.

```
import numpy as np
```

```
array=np.array([10,20,30,40,50])
```

```
print(array.ndim)
print(array.shape)
print(array.size)
print(array.dtype)
```
